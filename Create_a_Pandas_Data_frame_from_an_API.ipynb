####Step 1:Import pandas






```python
import pandas as pd
```



####Step 2: Import requests



```python
import requests
```


####Step 3: Make a GET Request from an API Endpoint


```python
url = 'https://api.covid19api.com/summary'
response = requests.get(url)
response
```



#### Step 4: Extract Data from the Request using the Json() Method

```python
# Weâ€™ll use a method called json() to extract the json-structured data from the request.

json = response.json()
json
```

#### Step 5: Exploring our JSON using the Keys() and Type() Methods

```python
#The keys() method returns a view object. The view object contains the keys of the dictionary, as a list.
json.keys()
```

```python
# Explore the Values of the keys
json['Global']
```

```python
json['Countries']
```

```python
# Check the type of each keys

print(type(json['Global']))
print(type(json['Countries']))
print(type(json['Date']))
```

### Step 6: JSON to Dataframe

```python
df = pd.DataFrame(json['Countries'])
df
```
